```python
import pandas as pd
df = pd.read_csv("../data/customer_support_tickets_dirty.csv")
```

```python
df.head()
```

```python
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df = df.drop_duplicates()
df = df.dropna(subset=["text", "label"])  # Using your actual column names
df.to_csv("../data/customer_support_tickets_clean.csv", index=False)
```
